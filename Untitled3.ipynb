```python
from sklearn.preprocessing import MinMaxScaler

import pandas as pd

df = pd.read_csv("train_dataset.csv")

df.head()

#  Kopya (orijinal veri bozulmasın)
df_norm = df.copy()

#  Hedef değişkeni ayırır
X = df_norm.drop("Is_Anemic", axis=1)
y = df_norm["Is_Anemic"]

# Scaler oluşturur
scaler = MinMaxScaler()

#  Normalizasyon uygula
X_scaled = scaler.fit_transform(X)

#  DataFrame'e geri çevir
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

#  Hedef değişkeni geri ekle
df_scaled = X_scaled.copy()
df_scaled["Is_Anemic"] = y.values

#  Kontrol et
df_scaled.head()
```
